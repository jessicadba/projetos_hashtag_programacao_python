#### Visualização de Dados simples e úteis: *head()*, *tail()*, *shape*, *decribe()*, *info()*:
São métodos do pandas usados para inspecionar um DataFrame e entender os dados rapidamente.
- **head()** - Retorna as primeiras linhas do DataFrame. Por padrão, exibe as 5 primeiras linhas, mas pode-se especificar quantas deseja ver.
- **tail()** - Retorna as últimas linhas do dataFrame. Por padrão, exibe as 5 últimas linhas.
- **shape** - Retorna a quantidade de linhas e colunas do DataFrame na forma de uma tupla *(linhas, colunas)*, sem precisar de parênteses porque é um atributo e não um método, sendo menos custoso em um DataFreme com muitos dados.
- **describe()** - Gera estatísticas descritivas sobre as colunas numéricas, como média, mínimo, máximo, desvio padrão, etc. Útil para entender a distribuição dos dados.

```python
display(vendas_df.head())
display(vendas_df.tail())
print(vendas_df.shape)
display(vendas_df.describe())

# Consultar informações básicas do DataFrame
vendas_df.info()
```

#### Método *.loc*
O método **.loc** (localizar) do pandas é muito importante, pois através dele é possível selecionar linhas e colunas de um DataFrame, com base nos rótulos (nomes de índices e colunas) e filtrar os dados de várias formas.
- Consultar 1 linha
- Consultar linhas de acordo com alguma condição
- Linhas e colunas específicas
- Consultar 1 valor específico

```python
# Consultar uma[1] ou mais linhas[1:5]
display(vendas_df.loc[1:5])

# Pegar linhas que correspondem a uma determinada condição exibindo o resultado com o display
display(vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping'])

# Armazenar o resultado da consulta em um DataFrame
vendas_norteshopping_df = vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping']

# Selecionar linhas e colunas do DataFrame usando loc. Ex sintaxe: vendas_df.loc[linhas, colunas]
vendas_norteshopping_df = vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping', ['ID Loja', 'Produto', 'Quantidade']]
display(vendas_norteshopping_df)

# Selecionar 1 valor específico
print(vendas_df.loc[1, 'Produto'])
```

#### Adicionando colunas em um DataFrame:
- Adicionando a partir de uma coluna que já existe;
- Criar uma coluna com valor padrão.

```python
# a partir de uma coluna que existe
vendas_df['Comissao'] = vendas_df['Valor Final'] * 0.05
#display(vendas_df)

# Criar uma coluna com valor padrão (lembrando a sintaxe loc.[linhas, colunas]

vendas_df['Imposto'] = 0 # não indicado para ser utilizados em Dataframes com grandes volumes de dados
display(vendas_df)

```

```python
# a partir de uma coluna que existe
vendas_df['Comissao'] = vendas_df['Valor Final'] * 0.05
display(vendas_df)

# Criar uma coluna com valor padrão:
# Forma NÃO indicada para ser utilizados em Dataframes com grandes volumes de dados.
vendas_df['Imposto'] = 0
display(vendas_df)

# Forma mais segura e menos custosa para criar colunas com um valor padrão é utilizar o loc.[linhas, colunas]
vendas_df.loc[:, 'Imposto'] = 0
display(vendas_df)

```

#### Adicionando linhas em um DataFrame:
- Linhas de um complemento da base de dados;
- O *pd.concat()* combina os dados do DataFrame;
- O argumento *ignore_index=True* garante que os índices sejam reorganizados corretamente.

```python
vendas_dez_df = pd.read_excel(r"C:\Users\13562080700\Desktop\treinamento\Vendas - Dez.xlsx")
#display(vendas_dez_df)

vendas_df = pd.concat([vendas_df, vendas_dez_df], ignore_index=True)
display(vendas_df)
```

#### Excluindo linhas e colunas de um DataFrame:

```python
# axis=1 - exclui colunas
# axis=0 - exclui linhas
#vendas_df = vendas_df.drop('Imposto', axis=1)
#display(vendas_df)

vendas_df = vendas_df.drop(108086, axis=0)
display(vendas_df)
```

#### Tratamento de valores nulos/vazios:
- Deletar linhas/colunas vazias;
- Deletar linhas que possuem valores vazios;
- Preencher valores vazios (média e último valor).

```python
# Para deletar as linhas e colunas vazias é utilizado o parâmetro *all*.
# Caso queira exluir apenas as colunas onde todos os valores estão vazios, deve-se informar o eixo (axis=1).
vendas_df.dropna(how='all')

# Deletar linhas que possuem pelo menos 1 valor vazio
vendas_df = vendas_df.dropna() #exclui as linhas que tem pelo menos um valor vazio (essa é a forma padrão de execução)

# Preencher valores vazios passando um valor
vendas_df['Comissao'] = vendas_df['Comissao'].fillna(0)

# Preencher valores vazios com a média da coluna
vendas_df['Comissao'] = vendas_df['Comissao'].fillna(vendas_df['Comissao'].mean())
display(vendas_df)

# Preencher com o último valor
vendas_df = vendas_df.ffill()
```

#### Calculo de Indicadores:
- Groupby
- Value Counts

```python
# Quantidade de transações por loja - Value Counts
transacoes_loja = vendas_df['ID Loja'].value_counts()
display(transacoes_loja)

# Faturamento de cada um dos produtos - Groupby (Quando agrupar a coluna Produto, o valor final será somado)
faturamento_produto = vendas_df[['Produto', 'Valor Final']].groupby('Produto').sum()
display(faturamento_produto)
```
